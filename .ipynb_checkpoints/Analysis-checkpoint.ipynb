```python
fraud_frame = df.groupby(['type', 'isFraud'])['type'].count().unstack()
fraud_frame
```

```python
fraud_frame = fraud_frame.fillna(0)
fraud_frame
```

```python
x_set = fraud_frame.index
y_set_0 = fraud_frame[0]
y_set_1 = fraud_frame[1]
#y_set_0

width = 0.4
values = np.arange(len(x_set))
```

```python
plt.bar(df['type'].unique(), df['type'].value_counts(), width)
plt.xlabel('Transaction type')
plt.ylabel('No. of Transactions')
plt.title('Transaction Types Graph')
plt.show()
```

```python
plt.bar(values - 0.2, y_set_0, width, color='green', label='No Frauds')
plt.bar(values + 0.2, y_set_1, width, color='red', label='Frauds')
plt.xticks(values, x_set)
plt.xlabel('Transaction type')
plt.ylabel('No. of Transactions')
plt.title('Transaction Types split by Fraud Graph')
plt.legend()
plt.show()
```

```python
cash_out_transactions = df.groupby('type').get_group('CASH_OUT')
#cash_out_transactions
```

```python
x_axis = cash_out_transactions['oldbalanceOrg']- cash_out_transactions['newbalanceOrig']
#x_axis
y_axis = cash_out_transactions['newbalanceDest']- cash_out_transactions['oldbalanceDest']
#y_axis
```

```python
plt.scatter(x_axis, y_axis, s=100)
plt.title('Scatter Plot for Cash Out Transaction')
plt.xlabel('Origin account balance Delta')
plt.ylabel('Destination account balance Delta')
plt.show()
```
